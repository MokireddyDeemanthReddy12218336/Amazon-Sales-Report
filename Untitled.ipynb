```python
df=df.drop(['New','PendingS'],axis=1)
```

```python
df.head()
```

```python
# Handling missing values in 'currency' and 'Amount'
df['currency'].fillna(df['currency'].mode()[0], inplace=True)  # Replace with mode
df['Amount'].fillna(df['Amount'].median(), inplace=True)  # Replace with median

# Handling missing values in 'ship-city', 'ship-state', 'ship-postal-code', 'ship-country'
for column in ['ship-city', 'ship-state', 'ship-postal-code', 'ship-country']:
    df[column].fillna(df[column].mode()[0], inplace=True)  # Replace with mode

# Handling missing values in 'fulfilled-by'
df['fulfilled-by'].fillna('Unknown', inplace=True)
```

```python
# Identifying duplicate rows 
duplicate_rows = df[df.duplicated()] 
print("\nDuplicate Rows:") 
print(duplicate_rows)
```

```python
df.dtypes
```

```python
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y', errors='coerce')
```

```python
order_counts = df['ship-state'].value_counts().head(10)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=order_counts.index, y=order_counts.values, palette='viridis')
plt.xlabel('Ship State')
plt.ylabel('Number of Orders')
plt.title('Top 10 Shipping States')
plt.xticks(rotation=90, fontsize=12)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Count the number of orders per Fulfilment
fulfilment_counts = df['Fulfilment'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(fulfilment_counts, labels=fulfilment_counts.index, autopct='%1.1f%%', colors=plt.get_cmap('Set2').colors)
plt.title('Distribution of Fulfilment')
plt.show()
```

```python
# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Amount'], bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Order Amount')
plt.grid(True)
plt.show()
```

```python
# Box plot for 'Amount' grouped by 'Category'
plt.figure(figsize=(8,6))
sns.boxplot(x='Category', y='Amount', data=df)
plt.title('Box Plot of Amount by Category')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Amazon Sale Report.csv')  
# Calculate the correlation matrix
corr = df.corr()
# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
# Set the title
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.violinplot(x='Status', y='Amount', data=df)
plt.title('Violin Plot of Amount by Status')
plt.show()
```

```python
# Prepare data for the area plot
df_area = df.groupby(['Date'])['Amount'].sum().reset_index()

# Plot area plot
plt.figure(figsize=(8,6))
plt.fill_between(df_area['Date'], df_area['Amount'], color="skyblue", alpha=0.4)
plt.plot(df_area['Date'], df_area['Amount'], color="Slateblue", alpha=0.6)
plt.title('Area Plot of Total Amount Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
# Density plot for 'Amount'
plt.figure(figsize=(8,6))
sns.kdeplot(df['Amount'], shade=True)
plt.title('Density Plot for Amount')
plt.show()
```

```python
df_grouped = df.groupby(['Category', 'Status']).size().unstack()

# Plot the stacked bar chart
ax = df_grouped.plot(kind='bar', stacked=True, figsize=(8,6))

# Set title
plt.title('Stacked Bar Plot of Status by Category')

# Move the legend outside the plot
plt.legend(title='Status', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.tight_layout()  # Adjust layout to make sure nothing is clipped
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Examining unique Values
def unique_value(data):
    unique = []
    for i in data.columns:
        x = data[i].value_counts().count()
        unique.append(x)
    return pd.DataFrame(unique, index = data.columns, columns = ['Total unique value'])
    
unique_value(df)
```

```python
# Create week and year columns in DataFrame
df['week'] = df['Date'].dt.isocalendar().week
df['year'] = df['Date'].dt.isocalendar().year

# Calculate the total Amount by week and year
weekly_amount = df.pivot_table(values='Amount', index=['year', 'week'], aggfunc='sum').reset_index()

# Create a 'Week' column to easily plot the data
weekly_amount['Week'] = pd.to_datetime(weekly_amount[['year', 'week']].astype(str).agg('-'.join, axis=1) + '-1', format='%Y-%W-%w')

# Plotting the data
plt.figure(figsize=(15, 10))
plt.plot(weekly_amount['Week'], weekly_amount['Amount'])
plt.title('Total Amount by Week')
plt.xlabel('Week')
plt.ylabel('Total Amount')
plt.grid()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Analysis output:
There is a significant increase from around March to May, followed by a gradual decline in subsequent weeks. 
This could suggest that there was a surge in sales activity during a certain period, possibly related to promotions 
or special events.

```python
# Group by Month and Category, summing the Amount
df['Month'] = df['Date'].dt.to_period('M')
monthly_category_amount = df.groupby(['Month', 'Category'])['Amount'].sum().reset_index()

#Plotting the data
plt.figure(figsize = (10, 8))
sns.barplot(x = 'Month', y = 'Amount', hue = 'Category', data = monthly_category_amount)
plt.title('Total Amount by Month and Category')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(rotation = 45)
plt.legend(title ='Category')
plt.grid()
plt.show()
```

Analysis output:
Shirts had the highest sales in March and May 2021.
Blazzers sales were relatively consistent across all months.
Shoes and Socks had lower sales compared to other categories.
Watches had moderate sales, peaking slightly in June 2021.
